```python
import matplotlib.pyplot as plt
import numpy as np

# TIME_CONVERTER = 0.2
# DIS_CONVERTER = 0.2358
```

```python
lateral_dis_rec_list = np.zeros((10, 499))
u_mid = np.zeros((10,499))
u_diff = np.zeros((10,499))
longitudinal_dis = np.zeros((10,499))
for i in range(10):
    cur_rec = np.load('./recdata' + str(i+1) + '.npy')
    lateral_dis_rec_list[i] = cur_rec[:499,0]
    u_mid[i] = cur_rec[:499,1]
    u_diff[i] =cur_rec[:499,2]
    longitudinal_dis[i] = cur_rec[:499,3]
```

```python
x = np.arange(0, 499, 1)
dis_mean_list = np.mean(lateral_dis_rec_list, axis=0)
dis_max_list = np.max(lateral_dis_rec_list, axis=0)
dis_min_list = np.min(lateral_dis_rec_list, axis=0)
plt.title('Mean Case')
plt.xlabel('time step')
plt.ylabel('lateral distance')
plt.plot(dis_mean_list)

plt.fill_between(x, dis_max_list, dis_min_list, color='grey', alpha='0.5')
```

```python
# x = np.arange(0, 499, 1)
u_mid_mean_list = np.mean(u_mid, axis=0)
u_mid_max_list = np.max(u_mid, axis=0)
u_mid_min_list = np.min(u_mid, axis=0)
plt.title('Mean Case')
plt.xlabel('time step')
plt.ylabel('u mid')
plt.plot(u_mid_mean_list)

plt.fill_between(x, u_mid_max_list, u_mid_min_list, color='grey', alpha='0.5')
```

```python
u_diff_mean_list = np.mean(u_diff, axis=0)
u_diff_max_list = np.max(u_diff, axis=0)
u_diff_min_list = np.min(u_diff, axis=0)
plt.title('Mean Case')
plt.xlabel('time step')
plt.ylabel('u diff')
plt.plot(u_diff_mean_list)

plt.fill_between(x, u_diff_max_list, u_diff_min_list, color='grey', alpha='0.5')
```

```python
longitudinal_dis_mean_list = np.mean(longitudinal_dis, axis=0)
longitudinal_dis_max_list = np.max(longitudinal_dis, axis=0)
longitudinal_dis_min_list = np.min(longitudinal_dis, axis=0)
plt.title('Mean Case')
plt.xlabel('time step')
plt.ylabel('longitudinal distance')
plt.plot(longitudinal_dis_mean_list)

plt.fill_between(x, longitudinal_dis_max_list, longitudinal_dis_min_list, color='grey', alpha='0.5')
```
