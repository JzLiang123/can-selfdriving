```python
import numpy as np
import matplotlib.pyplot as plt
```

# pure pursuit

```python
# sin_alpha_list = np.load('./pure-pursuit/record.npy')
# plt.plot(sin_alpha_list)
# plt.title('Test')
# plt.xlabel('time step')
# plt.ylabel('sin alpha')
dist_rec_list = np.zeros((10, 400))
red_rec_list = np.zeros((10, 400))
pwm_rec_list = np.zeros((10, 400))
for i in range(10):
    cur_rec = np.load('./pure-pursuit/record' + str(i+1) + '.npy')
    dist_rec_list[i] = cur_rec[:400, 0]
    red_rec_list[i] = cur_rec[:400, 1]
    pwm_rec_list[i] = cur_rec[:400, 2]
```

```python
#distance to tangent
for i, rec in enumerate(dist_rec_list):
    plt.plot(rec)
    plt.title('Test Case ' + str(i+1))
    plt.xlabel('time step')
    plt.ylabel('distance to tangent')
```

```python
dist_mean_pp = np.mean(dist_rec_list, axis=0)
dist_max_pp = np.max(dist_rec_list, axis=0)
dist_min_pp = np.min(dist_rec_list, axis=0)

# Plot lines
x = np.arange(0, 400, 1)
mean_line, = plt.plot(x, dist_mean_pp, label='mean')
# max_line, = plt.plot(x, dis_max_pi, label='max')
# min_line, = plt.plot(x, dis_min_pi, label='min')

# plt.legend(handles=[mean_line, max_line, min_line])
plt.fill_between(x, dist_max_pp, dist_min_pp, color='grey', alpha='0.5')
```

```python
#redian at tangent
for i, rec in enumerate(red_rec_list):
    plt.plot(rec)
    plt.title('Test Case ' + str(i+1))
    plt.xlabel('time step')
    plt.ylabel('radian at tangent')
```

```python
red_mean_pp = np.mean(red_rec_list, axis=0)
red_max_pp = np.max(red_rec_list, axis=0)
red_min_pp = np.min(red_rec_list, axis=0)

# Plot lines
x = np.arange(0, 400, 1)
mean_line, = plt.plot(x, red_mean_pp, label='mean')
# max_line, = plt.plot(x, dis_max_pi, label='max')
# min_line, = plt.plot(x, dis_min_pi, label='min')

# plt.legend(handles=[mean_line, max_line, min_line])
plt.fill_between(x, red_max_pp, red_min_pp, color='grey', alpha='0.5')
```

```python
#pwm difference
for i, rec in enumerate(pwm_rec_list):
    plt.plot(rec)
    plt.title('Test Case ' + str(i+1))
    plt.xlabel('time step')
    plt.ylabel('pwm difference')
```

```python
pwm_mean_pp = np.mean(pwm_rec_list, axis=0)
pwm_max_pp = np.max(pwm_rec_list, axis=0)
pwm_min_pp = np.min(pwm_rec_list, axis=0)

# Plot lines
x = np.arange(0, 400, 1)
mean_line, = plt.plot(x, pwm_mean_pp, label='mean')
# max_line, = plt.plot(x, dis_max_pi, label='max')
# min_line, = plt.plot(x, dis_min_pi, label='min')

# plt.legend(handles=[mean_line, max_line, min_line])
plt.fill_between(x, pwm_max_pp, pwm_min_pp, color='grey', alpha='0.5')
```
