```python
import matplotlib.pyplot as plt
import numpy as np

raw_value = np.load('record.npy')
yaw = raw_value[:, 0]
pitch = raw_value[:, 1]
roll = raw_value[:, 2]

plt.figure()
plt.plot(yaw)
plt.title('Yaw ')
plt.xlabel('time step')
plt.ylabel('degree')
```

```python
plt.figure()
plt.plot(pitch)
plt.title('Pitch ')
plt.xlabel('time step')
plt.ylabel('degree')
```

```python
plt.figure()
plt.plot(roll)
plt.title('Roll')
plt.xlabel('time step')
plt.ylabel('degree')
```
